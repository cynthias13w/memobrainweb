```python
import plotly.express as px
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
import pandas as pd
```

```python
# Import OASIS

oasis1 = pd.read_csv('oasis_cross-sectional.csv')
oasis1
```

```python
oasis1 = oasis1[oasis1.CDR.notna()]
oasis1
```

```python
x = oasis1.groupby('SES').count()['ID'].index
x
```

```python
y = oasis1.groupby('SES').count()['ID'].values
y
```

```python
sns.barplot(x, y, palette= 'magma')
plt.title('Barplot: Socio-economic status of participants')
plt.xlabel('Socio-economic status')
plt.ylabel('Count')
plt.savefig('oasis1_ses.png')
```

```python
sns.histplot(oasis1.Age, color= 'c', kde = True)
plt.title('Histogram: Age of participants')
plt.savefig('oasis1_age.png')
```

```python
plt.figure(figsize = (12, 6))
plt.title('Pie Chart: Education level of participants')
plt.pie(oasis1.groupby('Educ').count()['ID'].values, 
        labels = ['Less than High School', 'High School', 'Some College', 'College', 'Beyond College'], 
        autopct='%1.1f%%', startangle=90, colors = ['plum', 'c', 'salmon', '#8EB897', '#B7C3F3'])
plt.savefig('oasis1_education.png');
```
